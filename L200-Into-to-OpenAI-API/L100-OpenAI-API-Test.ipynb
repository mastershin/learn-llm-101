```python
# https://platform.openai.com/docs/api-reference
import os

def is_running_in_colab():
    try:
        import google.colab
        return True
    except ImportError:
        return False
    
def get_api_key(keyname):
    # Check if 'API_KEY' is in environment, or .env file
    if keyname not in os.environ:
        if is_running_in_colab():
            from google.colab import userdata
            return userdata.get(keyname)
        elif os.path.exists(".env"):
            from dotenv import load_dotenv

            load_dotenv(".env")
    return os.environ.get(keyname)

OPENAI_KEY = get_api_key("OPENAI_KEY")
if OPENAI_KEY:
    print("API KEY Loaded")
```

```python
# Import the OpenAI class from the openai module
from openai import OpenAI

# Create an instance of the OpenAI class with the provided API key
client = OpenAI(api_key=OPENAI_KEY)

# Set the user input to the string "Hi"
user_input = "Describe about your capability as an LLM in detail using the bullet lists."

# Create a chat completion using the OpenAI client and the user input
model_id = "gpt-3.5-turbo"
response = client.chat.completions.create(
    model=model_id, messages=[{"role": "user", "content": user_input}]
)

# Print the response from the chat completion
print(response.choices[0].message.content)
```
